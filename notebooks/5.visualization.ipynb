```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# scrape geo json from the web 
import requests
import json

def get_country_geojson(file_name, abreviation):
    url = f"https://geodata.ucdavis.edu/gadm/gadm4.1/json/gadm41_{abreviation}_0.json"
    r = requests.get(url)
    cont = r.json()
    with open(file_name, 'w') as outfile:
        outfile.write(json.dumps(cont))
```
